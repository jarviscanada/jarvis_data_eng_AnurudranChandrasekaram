```python
import pandas as pd
df_bookings = pd.read_csv('data/bookings.csv',delimiter=',',header='infer')
df_members = pd.read_csv('data/members.csv',delimiter=',',header='infer')
df_facilities = pd.read_csv('data/facilities.csv',delimiter=',',header='infer')
```

```python
# Retrieve the start times of members' bookings
pd.merge(df_members, df_bookings,
how='inner', on='memid').query('firstname=="David"').query('surname=="Farrell"').starttime
```

```python
#produce a list of the start times for bookings for tennis courts, for the date '2012-09-21'. Return a list of start time and facility name pairings, ordered by the time.
pd.merge(df_bookings, df_facilities,
how='inner', on='facid').query("name.str.startswith('Tennis Court')").query('starttime>="2012-09-21"').query('starttime<"2012-09-22"')
```

```python
#Produce a list of all members, along with their recommender, using no joins.
res = pd.merge(df_members, df_members,
how='left', left_on='recommendedby', right_on="memid")
res["member"] = res["firstname_x"] + " " + res["surname_x"]
res["recommender"] = res["firstname_y"] + " " + res["surname_y"]
res[["member", "recommender"]].sort_values(by=["member"])
```

```python
#Count the number of recommendations each member makes.
df_members.groupby(df_members.recommendedby).recommendedby.count()
```

```python
#List the total slots booked per facility
df_bookings.groupby(df_bookings.facid).slots.sum()
```

```python
#List the total slots booked per facility in a given month
df_bookings.query('starttime>="2012-09-01"').query('starttime<"2012-10-01"').groupby(df_bookings.facid).slots.sum().sort_values(ascending=True)
```

```python
#List the total slots booked per facility per month
df_bookings.loc[df_bookings["starttime"].str.contains("2012")].groupby(['facid',df_bookings.loc[df_bookings["starttime"].str.contains("2012")].starttime.str[5:7]]).slots.sum()

```

```python
#Find the count of members who have made at least one booking
df_bookings["memid"].unique().size
```

```python
#List each member's first booking after September 1st 2012
pd.merge(df_members, df_bookings,
how='inner', on='memid').query('starttime>="2012-09-01"')[['starttime', 'firstname', 'surname', 'memid']].groupby(["surname", "firstname", "memid"]).min().sort_values(by=['memid'])
```

```python
#Perform a case-insensitive search
df_facilities.loc[df_facilities["name"].str.contains('^tennis', case=False)]
```
